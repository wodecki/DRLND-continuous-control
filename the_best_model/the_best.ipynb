```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cols = ['model', 'episode', 'avg. score', 'score', 'UPD', 'BUFF_SIZE','BATCH_SIZE', 'fc1', 'fc2', \
        'LR_ACTOR', 'A_gradient_clip', 'A_leaky', 'A_dropout',\
       'LR_CRITIC', 'C_gradient_clip', 'C_BATCH_NORM','C_leaky', 'C_dropout']
```

```python
exp2 = pd.read_csv('results.txt', names = cols)
```

```python
fig = plt.figure(figsize=(8, 7))
x = exp2[exp2.model == 14]['episode']
y = exp2[exp2.model == 14]['avg. score']
ys = exp2[exp2.model == 14]['score']
plt.plot(x, y)
plt.plot(x, ys)
plt.legend(loc='upper right')
plt.title("The best model")
plt.xlabel('Episode number')
plt.ylabel('(avg) score')
plt.show()
```
