```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cols = ['model', 'episode', 'avg. score', 'score', 'UPD', 'BUFF_SIZE','BATCH_SIZE', 'fc1', 'fc2', \
        'LR_ACTOR', 'A_gradient_clip', 'A_leaky', 'A_dropout',\
       'LR_CRITIC', 'C_gradient_clip', 'C_BATCH_NORM','C_leaky', 'C_dropout']
```

```python
exp1 = pd.read_csv('results.txt', names = cols)
```

```python
fig = plt.figure(figsize=(8, 7))
for i in range(1, 17):
    x = exp1[exp1.model == i]['episode']
    y = exp1[exp1.model == i]['avg. score']
    #lab = 'model = '+str(i)
    plt.plot(x, y, label = 'model no. '+ str(i))
    plt.legend(loc='upper right')
plt.title("Experiment 1: results")
plt.xlabel('Episode number')
plt.ylabel('Avg score')
plt.show()
```
